```python
import tensorflow as tf
import numpy as np
import IPython.display as ipd
import soundfile as sf
from e2e_tts import load_model,e2e_tts
```

```python
%%time
%%capture
tacotron2,fastspeech2,mb_melgan = load_model()
```

```python
input_text = "有線電視大豐（6184）去年每股純益3.38元，董事會決議擬派發現金股利3元，集團連續兩年合併營收突破20億元，主要動能就是來自寬頻用戶數的成長。"
```

```python
%%time
%%capture
uuid = e2e_tts(input_text, tacotron2,fastspeech2,mb_melgan)    
```

## tacotron

```python
%%time
tacotron_path = './output/' + uuid + '_tacotron.wav'
audio1,_ = sf.read(tacotron_path)
ipd.Audio(audio1, rate=24000)
```

## fastspeech

```python
%%time
fastspeech_path = './output/' + uuid + '_factspeech.wav'
audio2,_ = sf.read(fastspeech_path)
ipd.Audio(audio2, rate=24000)
```
